## Multi-Object Tracking with Ultralytics YOLO

```python

from ultralytics import YOLO

# Load an official or custom model
model = YOLO('yolov8n.pt')  # Load an official Detect model


# Perform tracking with the model
#results = model.track(source="2.mp4", save=True)  # Tracking with default tracker (The default tracker is BoT-SORT.)
results = model.track(source="c.jpg", save=True) 
#results = model.track(source="https://youtu.be/LNwODJXcvt4", show=True, tracker="bytetrack.yaml")  # Tracking with ByteTrack tracker
```

```python
from ultralytics import YOLO

# Configure the tracking parameters and run the tracker
model = YOLO('yolov8n.pt')

results = model.track(source="1.mp4",conf=0.7, iou=0.5, save=True)  
```

```python
from ultralytics import YOLO

# Load an official or custom model
model = YOLO('yolov8n-seg.pt')  # Load an official Segment model

# Perform tracking with the model
#results = model.track(source="HeartofStoneNetflix.mp4", save=True) # Tracking with default tracker
results = model.track(source="1.mp4", save=True, tracker="bytetrack.yaml")  # Tracking with ByteTrack tracker
```

```python
from ultralytics import YOLO

# Load an official or custom model
model = YOLO('yolov8n-pose.pt')  # Load an official Pose model

# Perform tracking with the model
results = model.track(source="1.mp4", save=True) # Tracking with default tracker
#results = model.track(source="https://youtu.be/LNwODJXcvt4", show=True, tracker="bytetrack.yaml")  # Tracking with ByteTrack tracker
```
